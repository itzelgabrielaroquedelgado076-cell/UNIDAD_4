**3.Diseo experimental para un factor**
El diseo experimental para un factor se utiliza cuando se desea evaluar si un solo factor categrico influye significativamente en una variable cuantitativa. En este proyecto, cada variable categrica se analiza por separado para determinar si genera diferencias en el puntaje del examen. Este enfoque permite comparar grupos (por ejemplo, elevado, medio y bajo nivel o rango de participacin parental) y evaluar si las diferencias entre ellos son estadsticamente significativas.

**4.ANOVA**
El ANOVA es una tcnica estadstica que permite comparar las medias de varios grupos para verificar si existen diferencias significativas entre ellos. En este caso se utiliza para analizar si el puntaje del examen vara entre los niveles de cada variable categrica. Un valor p menor a 0.05 indica que al menos un grupo presenta un puntaje distinto.

**5.Prueba de Tukey**
Cuando el ANOVA indica que existen diferencias significativas, la prueba de Tukey permite identificar entre qu grupos ocurren esas diferencias. Es una prueba post hoc utilizada para comparar todas las parejas de medias, sealando especficamente cules categoras presentan diferencias en sus puntajes. Esto ayuda a interpretar de manera ms clara el comportamiento de cada variable categrica.

**6.Eta cuadrado como medida del tamao del efecto**
El eta cuadrado se emplea para determinar qu tan importante es la diferencia encontrada. Aunque el ANOVA puede detectar diferencias estadsticamente significativas, el eta cuadrado indica si estas diferencias tienen una relevancia prctica. Valores cercanos a 0 reflejan efectos pequeos, mientras que valores ms altos indican que la variable categrica explica una mayor proporcin de la variabilidad en el puntaje del examen.

La interpretacin del eta cuadrado se basa en criterios comnmente aceptados para medir el tamao del efecto:

-0.01  < 0.06  Efecto pequeo

-0.06  < 0.14  Efecto mediano
La variable explica una proporcin moderada de la variabilidad.

  0.14  Efecto grande
La variable tiene un impacto importante en el puntaje de examen.

Esto nos permiti complementar el anlisis estadstico, distinguiendo no solo si una variable generaba diferencias significativas, sino tambin qu tan importante era su efecto sobre el puntaje de examen.
```python
import pandas as pd

url="https://raw.githubusercontent.com/itzelgabrielaroquedelgado076-cell/UNIDAD_4/refs/heads/main/dataset.csv"

df=pd.read_csv(url)
df= df.dropna()

df
```

```python
from statsmodels.formula.api import ols
import statsmodels.api as sm
#Y~X
modelo_lineal = ols(
    'puntaje_examen ~ horas_estudio + asistencia + C(participacion_parental) + C(acceso_a_recursos) + C(actividades_extracurriculares) + horas_suenio + calificaciones_previas + C(nivel_motivacion) + C(acceso_internet) + sesiones_tutoria + C(ingreso_familiar) + C(calidad_profesorado) + C(tipo_escuela) + C(influencia_companieros) + actividad_fisica + C(discapacidad_aprendizaje) + C(nivel_estudio_padres) + C(distancia) + C(genero)',
    data=df).fit()
tabla_anova = sm.stats.anova_lm(modelo_lineal)
tabla_anova
#Itzel Gabriela Roque Delgado
```

**CONCLUSIN:**

Las variables ms relevantes que influyen significativamente en el puntaje final del examen son la participacin parental, acceso a recursos, actividades extracurriculares, nivel de motivacin del estudiante, acceso a internet, ingreso familiar, calidad profesorado, influencia de compaeros, discapacidad de aprendizaje, nivel de estudio de padres, distancia, horas de estudio, asistencia, calificaciones previas, sesiones de tutoria y la actividad fsica debido a que el valor-p de cada una de estas variables es menor al nivel de significancia por lo que se rechaza la hiptesis nula y nos indica que si hay diferencia significativa; mientras que las variables que no influyen significativamente son el tipo de escuela, genero y horas de sueo debido a que estas variables tienen un valor-p mayor al nivel de significancia de 0.05 lo que en este caso nos indica que no se rechaza la hiptesis nula por lo que no hay diferencia significativa.  

```python
df["participacion_parental"].unique()
```

```python
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import matplotlib.pyplot as plt

nivel_de_significancia = 0.05

# Prueba de Tukey
tukey = pairwise_tukeyhsd(
    endog=df['puntaje_examen'], #variable de respuesta
    groups=df['participacion_parental'], #variable categorica
    alpha=nivel_de_significancia) #nivel de significancia

# Mostrar los resultados
print(tukey)

# Grfico de las diferencias entre grupos
tukey.plot_simultaneous(ylabel="Participacin parental", xlabel="Puntaje de examen final")

plt.gca().spines['right'].set_visible(False) # derecha
plt.gca().spines['top'].set_visible(False)   # superior
plt.title("Mltiples comparaciones entre todos los pares (Tukey)")

#Muestra los intervalos o niveles de confianza en las lineas, lo cual segn el grfico dentro de la variable participacin parental
#el rango de participacin que tiene un porcentaje de examen final ms elevadoes elevado
#Conforme a la prueba de tukey si existe o si hay diferencia significativa entre los grupos.
```

```python
#  (eta cuadrado)	Tamao del efecto
# ~0.01	              Pequeo
# ~0.06	              Mediano
# ~0.14 o ms	        Grande

# Se interpreta como el coeficiente de determinacin

# Estos valores pueden variar segn la disciplina. En estudios de ciencias
# sociales, un  de 0.06 puede ser muy significativo;
# en fsica o ingeniera, quiz se espera ms.

eta_squared = tabla_anova['sum_sq']['C(participacion_parental)'] / tabla_anova['sum_sq'].sum()
print(f" (tamao de efecto): {eta_squared:.3f}")
#Itzel Gabriela Roque Delgado
```

**CONCLUSIN:**

La participacin parental es el grado en que los padres o tutores se involucran en la educacin y formacin de sus hijos, tanto en casa como en la escuela. Los resultados de la prueba de Tukey indican que existen diferencias significativas en el puntaje del examen segn el nivel de participacin parental, ya que todos los grupos de comparacin que abarca el bajo y elevado, bajo y medio, y elevado y medio muestran diferencias estadsticamente significativas ya que el valor p ajustado de cada uno es menor al nivel de significancia de 0.05 por lo que se acepta la hipotesis alternativa; en el que en particular, los estudiantes con participacin parental elevada obtienen los puntajes de examen final promedio ms elevados, seguidos por quienes cuentan con una participacin media, mientras que la participacin baja se asocia con los puntajes ms bajos. Sin embargo, aunque la diferencia es estadsticamente significativa, el valor del eta cuadrado = 0.025 indica que el 2.5% de la variabilidad en el puntaje de examen final se puede explicar por la participacin parental, lo que representa un efecto pequeo; esto significa que, si bien el nivel de apoyo de los padres est relacionado con un mejor rendimiento acadmico, pero no es un factor determinante por s solo y la mayor parte de la variabilidad del puntaje se debe a otros factores que se encuentran dentro del 97.5% restante.
